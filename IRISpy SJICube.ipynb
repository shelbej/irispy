```python
import sunpy.map

from irispy.sji import SJICube
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.animation as animation
from astropy.coordinates.sky_coordinate import SkyCoord as SC
import astropy.units as u
from IPython.display import HTML
%matplotlib inline
i=8
```

```python
from irispy.data import sample
#
sji1330=SJICube(sample.SJI_CUBE_1330)
sji1400=SJICube(sample.SJI_CUBE_1400)
sji2796=SJICube(sample.SJI_CUBE_2796)
sji2832=SJICube(sample.SJI_CUBE_2832)
```

```python
print(sji1330.mean().min(),sji1330.mean().mean(),sji1330.mean().std())
sji1330norm = colors.Normalize(-32010, -31700)
sji1330map.plot_settings['norm']=sji1330norm
sji1330map.plot()
```

```python
sji1400norm = colors.Normalize(-31970, -31500)
sji1400.mean().plot_settings['norm']=sji1400norm
sji1400.max().plot_settings['norm']=sji1400norm
sji1400.std().plot_settings['norm']=sji1400norm
sji1400.max().peek()
sji1400.mean().peek()
sji1400.std().peek()
```

```python
maplist=[]
fig=plt.figure(dpi=100)
ax0=fig.add_axes([0,0,1,1])
sji2796norm = colors.Normalize(-31710, -30200)
for n in sji2796:
    n.plot_settings['norm']=sji2796norm
    maplist.append([n.plot(ax0)])
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.ArtistAnimation(fig, maplist, interval=20, blit=True)
HTML(anim.to_html5_video())
#anim.save()
```

```python
# Add overlay slit
sji2832[i].draw_slit()
# Add a overlay grid.
sji2832[i].draw_grid(grid_spacing=5*u.deg)
# Define a region of interest
length = 50 * u.arcsec
x0 = 720 * u.arcsec
y0 = -150 * u.arcsec

sji2832norm = colors.Normalize(-30900, -29100)
sji2832.max().plot_settings['norm']=sji2832norm
sji2832.max().plot()
```

```python
bl = [SC(775 * u.arcsec, -155 * u.arcsec, frame= 'helioprojective'),
      SC(800 * u.arcsec, -155 * u.arcsec, frame= 'helioprojective'),
      SC(825 * u.arcsec, -155 * u.arcsec, frame= 'helioprojective'),
      SC(850 * u.arcsec, -155 * u.arcsec, frame= 'helioprojective')]
tr = [SC(800 * u.arcsec, -55 * u.arcsec, frame= 'helioprojective'),
      SC(825 * u.arcsec, -55 * u.arcsec, frame= 'helioprojective'),
      SC(850 * u.arcsec, -55 * u.arcsec, frame= 'helioprojective'),
      SC(875 * u.arcsec, -55 * u.arcsec, frame= 'helioprojective')]

submap0=sji2832[i].submap(bl[0],tr[0])
submap1=sji2796[i].submap(bl[1],tr[1])
submap2=sji1330[i].submap(bl[2],tr[2])
submap3=sji1400[i].submap(bl[3],tr[3])
```

```python
fig=plt.figure(dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.set_xlim(775,875)
submap0.plot_settings['norm']=sji2832norm
submap0.plot(ax)
submap1.plot_settings['norm']=sji2796norm
submap1.plot(ax)
submap2.plot_settings['norm']=sji1330norm
submap2.plot(ax)
submap3.plot_settings['norm']=sji1400norm
submap3.plot(ax)
#submap1.draw_slit()

ax.set_title('IRIS SJI Sample Data')
ax.annotate(submap0.wavelength._repr_latex_(),xy=(bl[0].data._lon.arcsec,bl[0].data._lat.arcsec),color='white')
ax.annotate(submap1.wavelength._repr_latex_(),xy=(bl[1].data._lon.arcsec,bl[1].data._lat.arcsec),color='white')
ax.annotate(submap2.wavelength._repr_latex_(),xy=(bl[2].data._lon.arcsec,bl[2].data._lat.arcsec),color='white')
ax.annotate(submap3.wavelength._repr_latex_(),xy=(bl[3].data._lon.arcsec,bl[3].data._lat.arcsec),color='white')
```
